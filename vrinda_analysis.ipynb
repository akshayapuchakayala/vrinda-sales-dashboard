```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../dataset/Uncleaned Vrinda Store Sales Data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
print(f'Clean shape: {df.shape}')
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('-', '_')
)
```

```python
df.info()
```

```python
df['amount'] = df['amount'].fillna(df['amount'].median())
```

```python
df['date'] = pd.to_datetime(df['date'])
df.info()
```

```python
# Standardise gender (M/Men → Men, W/Women → Women)
df['gender'] = df['gender'].replace({'M':'Men', 'W':'Women'})
```

```python
df['month'] = df['date'].dt.month_name()
df['year'] = df['date'].dt.year
# Amount to numeric
df['amount'] = pd.to_numeric(df['amount'], errors='coerce')
```

```python
df.to_csv("../cleaned_data/cleaned_vrinda_sales.csv", index=False)
print('Clean file saved → cleaned_vrinda_sales.csv')
```

```python

total_revenue = df['amount'].sum()

print("Total Revenue:", total_revenue)
```

```python
#Total Orders
total_orders = df['order_id'].nunique()

print("Total Orders:", total_orders)
```

```python
#Average Order Value
aov = df['amount'].sum() / df['order_id'].nunique()

print("Average Order Value:", round(aov,2))
```

```python
#Monthly Sales Analysis
monthly_sales = df.groupby('month')['amount'].sum().sort_values()

print(monthly_sales)
```

```python
monthly_sales.plot(kind='bar', color='orange')

plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales Amount")

plt.show()
```

```python
#Sales by Gender
gender_sales = df.groupby('gender')['amount'].sum()

print(gender_sales)
```

```python
sns.countplot(x='gender', data=df)

plt.title("Number of Orders by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

```python
#Order Status Analysis
df['status'].value_counts()

```

```python
#Top 10 States by Sales
top_states = df.groupby('ship_state')['amount'].sum().nlargest(10)

print(top_states)
```

```python
#Sales Channels Performance
channel_sales = df.groupby('channel')['amount'].sum().sort_values(ascending=False)

print(channel_sales)
```

```python
channel_sales.plot(kind='bar', color='red')

plt.title("Sales by Channel")
plt.xlabel("Channel")
plt.ylabel("Sales")

plt.show()
```

```python
#Age Group Analysis
#Create Age Groups
bins = [0,18,25,35,45,55,100]

labels = ['0-18','19-25','26-35','36-45','46-55','55+']

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

age_sales = df.groupby('age_group',observed=False)['amount'].sum()

print(age_sales)
```

```python
age_sales.plot(kind='bar')

plt.title("Sales by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Sales Amount")

plt.show()
```

```python
#Top 10 Products
top_products = df.groupby('sku')['amount'].sum().nlargest(10)

print(top_products)
```

```python
top_products.plot(kind='bar')

plt.title("Top 10 Products")
plt.xlabel("Product SKU")
plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.show()
```

```python
# =========================
# KPI METRICS
# =========================
```

```python
#Total Revenue
#Meaning : Total money generated from all sales.

total_revenue = df['amount'].sum()

print("Total Revenue:", total_revenue)
```

```python
#Total Orders
#Meaning:How many unique orders exist.
total_orders = df['order_id'].nunique()

print("Total Orders:", total_orders)
```

```python
#Average Order Value (AOV)
#Meaning : Average money spent per order.
aov = total_revenue / total_orders

print("Average Order Value:", round(aov,2))
```

```python
#Top State
#Meaning : State generating highest sales.

top_state = df.groupby('ship_state')['amount'].sum().idxmax()

print("Top State:", top_state)
```

```python
#Top Category
#Meaning : Best selling category.

top_category = df.groupby('category')['amount'].sum().idxmax()

print("Top Category:", top_category)
```

```python
#Top Sales Channel
#Use your actual column name.

top_channel = df.groupby('channel')['amount'].sum().idxmax()

print("Top Channel:", top_channel)
```

```python
# Return Rate
# Meaning : Percentage of returned orders.

returned_orders = df[df['status'] == 'Returned']['order_id'].count()

return_rate = (returned_orders / total_orders) * 100

print("Return Rate:", round(return_rate,2), "%")
```

```python
# Cancellation Rate
cancelled_orders = df[df['status'] == 'Cancelled']['order_id'].count()

cancellation_rate = (cancelled_orders / total_orders) * 100

print("Cancellation Rate:", round(cancellation_rate,2), "%")
```

```python
# Sunnary table
kpi_summary = pd.DataFrame({
    'KPI': [
        'Total Revenue',
        'Total Orders',
        'Average Order Value',
        'Return Rate',
        'Cancellation Rate'
    ],
    
    'Value': [
        total_revenue,
        total_orders,
        round(aov,2),
        round(return_rate,2),
        round(cancellation_rate,2)
    ]
})

print(kpi_summary)
```
